# T5.1 (Correto)

Nesta questão vamos utilizar o Método dos Trapézios para determinar o valor numérico da integral abaixo, com pelo menos oito casas decimais exatas.

$$
  \int_{a}^{b} f(x) \,dx = \int_{1.2}^{2} \frac{e^x}{x} \,dx
$$

Para isso, divida o intervalo de integração em $n=2,3,4,5,\cdots$ subintervalos de mesmo tamanho $h=\frac{b−a}{n}$ e, para cada uma dessas subdivisões, estime o valor numérico da integral pela fórmula

$$
  T(n)=\frac{h}{2}(y_0+2y_1+2y_2+\cdots+2y_n−1+y_n)
$$

Utilize como critério de parada o menor n para o qual os valores $T(n−1)$ e $T(n)$ coincidem nas primeiras oito casas decimais.

# Resolução

A fórmula para calcularmos integrais numericamente pelo método dos trapézios é:

$$
\sum_{i=0}^{n−1}h⋅f(x_i)+f(x_i+1) = \frac{h}{2}\cdot[f(x_0)+2f(x_1)+2f(x_2)+\cdots+2f(x_{n−1})+f(x_n)]
$$

```octave
a = 1.2;
b = 2;
f = @(x) e^x/x;
n = 300;

[r] = trapezios(a,b,f,n);
```
